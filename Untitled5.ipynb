# **Hand Written Digit Prediction- Classification Analysis**

The Digits dataset consist of 8x8 pixel of digit. The images attributes of the dataset store 8x8 array of grayscales values for each image. We will use these arrays to visualize the images. The target attribute of dataset stores the digit each image represents

Import Library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Import Data

```python
from sklearn.datasets import load_digits
```

```python
df=load_digits()
```

```python
_, axes=plt.subplots(nrows=1,ncols=4,figsize=(10,3))
for ax,image,label in zip(axes,df.images,df.target):
  ax.set_axis_off()
  ax.imshow(image,cmap=plt.cm.gray_r,interpolation="nearest")
  ax.set_title("Training: %i"% label)
```

# **Data Preprocessing**

```python
n_samples=len(df.images)
data=df.images.reshape((n_samples,-1))
```

# **Scaling Image Data**

```python
data.min()
```

```python
data.max()
```

```python
data=data/16
```

Train Test Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test=train_test_split(data,df.target,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
y_pred
```

model accuracy

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
